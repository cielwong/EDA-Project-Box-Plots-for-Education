```python
list_col = []
for col in df.columns: 
    list_col.append(col)
del list_col[0]
list_col.remove('FTE')
list_col.remove('Total')
list_col
```

```python
list_num = ['FTE','Total']
list_num
```

```python
df2 = df.loc[:,list_num]
df2
```

```python
df3 = df.loc[:,list_col]
df3
```

```python
df3 = df3.fillna("empty")
df3
```

```python
counter = 0
freqtable = pd.DataFrame (columns = ['Value','Frequency'])
list_len = list()
for col in list_col:
    temp = df3[col].value_counts()
    temp = temp.to_frame()
    temp = temp.reset_index()
    temp.columns = ['Value','Frequency']
    list_temp = temp.index
    a = int(len(list_temp))+counter
    list_len.append(a)
    counter = counter + len(list_temp)
    freqtable = freqtable.append(temp)
freqtable = freqtable.reset_index()
freqtable=freqtable.drop('index', axis=1)
freqtable
```

```python
list_len
```

```python
for i in range(len(freqtable['Value'])):
        freqtable.iloc[i,0] = freqtable.iloc[i,0].strip()
freqtable
```

```python
warnings.filterwarnings('ignore')
counter = 0
cap = [14500,10000,8000,6000,6000,6000,6000,6000,6000,8000,6000,10000,8000,6000,6000,8000,8000,6000,6000,12000,7000,7000,6000]
for i in range(len(list_len)):
    list_temp = freqtable.iloc[counter:list_len[i]]
    counter = list_len[i]
    for j in range(len(list_temp)):
        if list_temp.iloc[j,1] < cap[i]: 
            list_temp.iloc[j,0] = "others"
    if sum(list_temp['Value'] == 'others') > 1:
        list_temp=list_temp.set_index('Value')
        list_temp.loc['others','Frequency']=sum(list_temp.loc['others','Frequency'])
        list_temp = list_temp.drop_duplicates()
        list_temp=list_temp.reset_index()
    plt.pie(list_temp["Frequency"], labels = list_temp["Value"] ,autopct='%1.1f%%')
    plt.title(list_col[i])
    plt.axis('equal')
    plt.show()
```
